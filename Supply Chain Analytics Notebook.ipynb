## **Ordered Based Analysis**

### **<span style="color: var(--vscode-foreground);">Calculate the total number of orders, orders on time, orders in full, and OTIF (on-time and in full) orders for each city.&nbsp;</span>**

```sql
WITH city_order_data AS (
    SELECT 
        dim_customers.city,
        fact_orders_aggregate.order_id,
        fact_orders_aggregate.on_time,
        fact_orders_aggregate.in_full,
        fact_orders_aggregate.otif
    FROM fact_orders_aggregate 
		INNER JOIN dim_customers 
			ON fact_orders_aggregate.customer_id = dim_customers.customer_id
),
all_order_data AS (
    SELECT 	
        city,
        COUNT(DISTINCT order_id) AS total_orders,
        SUM(CASE WHEN on_time = 1 THEN 1 ELSE 0 END) AS total_on_time,
        SUM(CASE WHEN in_full = 1 THEN 1 ELSE 0 END) AS total_in_full,
        SUM(CASE WHEN otif = 1 THEN 1 ELSE 0 END) AS total_otif
    FROM city_order_data
    GROUP BY city
)

SELECT 
    all_order_data.city,
    all_order_data.total_orders,
    all_order_data.total_on_time,
    all_order_data.total_in_full,
    all_order_data.total_otif,
    (SELECT COUNT(DISTINCT order_id) FROM fact_orders_aggregate) AS overall_total_order
FROM all_order_data;
```

### **Explanation:**

The first CTE, named `city_order_data`, combines the `fact_orders_aggregate` table with the `dim_customers` table using the `customer_id` column. This operation retrieves the city information associated with each order. The selected columns include `city`, `order_id`, `on_time`, `in_full`, and `otif`.

The second subquery, named `all_order_data`, groups the data from the `city_order_data` CTE by the `city` column. It calculates aggregate values such as the total number of orders, on-time orders, in-full orders, and OTIF orders for each city.

**Main Query:** The main query builds upon the `all_order_data` CTE and selects pertinent information. It retrieves the city name `city`, the total number of orders `total_orders`, the total number of on-time orders `total_on_time_orders` the total number of in-full orders `total_in_full_orders`, and the total number of OTIF orders `total_otif_orders`. Additionally, it counts all the orders from the `fact_orders_aggregate` table to obtain the `overall_total_order`.

This query will give you the total number of orders, total number of on-time orders, total number of in-full orders, and total number of OTIF orders for each city, along with the overall total count of all orders.

## **Analyzing Delivery Performance**

### **Provide insights into the distribution of the above metrics by customers within each city.**

```sql
WITH customer_metrics AS (
	SELECT 
		c.customer_name,
        SUM(ol.order_qty) AS total_orders,
        SUM(CASE WHEN o.on_time = 1 THEN ol.order_qty ELSE 0 END) AS total_orders_on_time,
        SUM(CASE WHEN o.in_full = 1 THEN ol.order_qty ELSE 0 END) AS total_orders_in_full,
        SUM(CASE WHEN o.otif = 1 THEN ol.order_qty ELSE 0 END) AS total_orders_otif
     FROM fact_order_lines AS ol
	 INNER JOIN dim_customers AS c  ON ol.customer_id = c.customer_id 
     INNER JOIN fact_orders_aggregate AS o ON ol.order_id = o.order_id 
     GROUP BY c.customer_name
) 
 
SELECT 
	customer_name,
    total_orders,
    total_orders_on_time,
    total_orders_in_full,
    total_orders_otif,
    ROUND((total_orders_on_time*100.0)/total_orders, 2) AS 'on_time_%',
    ROUND((total_orders_in_full*100.0)/total_orders, 2) AS 'in_full_%',
    ROUND((total_orders_otif*100.0)/total_orders, 2) AS 'otif_%'
FROM customer_metrics 
ORDER BY total_orders DESC;
```

### **Explanation:**

This query employs a Common Table Expression (CTE) named "customer\_metrics" to calculate essential order metrics for each customer.

The CTE combines data from the `fact_order_lines` and `fact_orders_aggregate` tables based on shared identifiers. It groups results by `customer_name`, aggregating order quantities to calculate the total number of orders, orders on time, in full orders, and OTIF orders.

Within the main query, three percentages—`on_time_percentage`, `in_full_percentage`, and `otif_percentage`—are calculated by dividing respective metrics by the `total_orders`.

The main query selects key metrics—`customer_name`, `total_orders`, `total_orders_on_time`, `total_orders_in_full`, `total_orders_otif`, `on_time_%`, and `in_full_%` and `otif_%`— from the CTE. The results are ordered in descending order of `total_orders`, showcasing top customers by order count.

This query provides actionable insights into customer performance and delivery metrics and a way to see which customers have the highest share of total orders, total orders on time, total orders in full, and total orders on time and in full (OTIF), and also the percentage of these metrics for each customer. This information can be used to identify which customers are performing well in terms of delivery performance, and which customers may need more attention.

### **Calculate the percentage variance between actual and target values for OT, IF, and OTIF metrics by city.**

```sql
 WITH actual AS (
	SELECT 
		dim_customers.city,
        SUM(CASE WHEN fact_orders_aggregate.on_time = 1 THEN 1 ELSE 0 END)* 100.0 / COUNT(DISTINCT fact_orders_aggregate.order_id)  AS actual_ot,
        SUM(CASE WHEN fact_orders_aggregate.in_full = 1 THEN 1 ELSE 0 END)* 100.0/ COUNT(DISTINCT fact_orders_aggregate.order_id)  AS actual_if,
        SUM(CASE WHEN fact_orders_aggregate.otif = 1 THEN 1 ELSE 0 END) * 100.0/ COUNT(DISTINCT fact_orders_aggregate.order_id) AS actual_otif
      FROM fact_orders_aggregate
      JOIN dim_customers ON fact_orders_aggregate.customer_id = dim_customers.customer_id 
      GROUP BY dim_customers.city
),
 target AS (
	SELECT 
		dim_customers.city,
        SUM(dim_targets_orders.ontime_target_pct) / COUNT(DISTINCT dim_targets_orders.customer_id) AS target_ot,
        SUM(dim_targets_orders.infull_target_pct)/ COUNT(DISTINCT dim_targets_orders.customer_id) AS target_if,
        SUM(dim_targets_orders.otif_target_pct) / COUNT(DISTINCT dim_targets_orders.customer_id) AS target_otif
	FROM dim_targets_orders
	JOIN dim_customers ON dim_targets_orders.customer_id = dim_customers.customer_id
	GROUP BY dim_customers.city
)
SELECT 
	actual.city,
	ROUND((actual.actual_ot - target.target_ot) * 100.0 / target.target_ot, 3) AS ot_varience,
	ROUND((actual.actual_if - target.target_if) * 100.0 / target.target_if, 3) AS if_varience,
	ROUND((actual.actual_otif - target.target_otif) * 100.0/ target.target_otif, 3) AS otif_varience
FROM actual
JOIN target ON actual.city = target.city;
```

### **Explanation:**

This query uses CTEs, joins, and aggregate functions to determine the variance between actual and target performance metrics for on-time (OT), in-full (IF), and on-time and in-full (OTIF) deliveries, categorized by city.

In the "actual" CTE, actual OT, IF, and OTIF delivery percentages are computed by dividing the sum of on-time(OT) and in-full(IF) and OTIF deliveries by the total number of distinct order IDs. The CTE merges the `fact_orders_aggregate` and `dim_customers` tables based on `customer_id` and groups outcomes by city.

The "target" CTE calculates targeted OT, IF, and OTIF delivery percentages by summing corresponding target values and dividing by the count of unique customer IDs. This CTE merges the `dim_targets_orders` and `dim_customers` tables based on `customer_id` and groups results by city.

The final SELECT statement unites the "actual" and "target" CTEs through a JOIN on the city column. The ROUND() function is employed to compute the percentage variance between actual and target performance for each metric. The query yields the city name and the variances for on-time, in-full, and on-time and in-full deliveries.

### **Identify the top and bottom 5 customers based on total quantity ordered, in full quantity ordered, and OTIF quantity ordered.**

```sql
-- Top 5 Cusotmers by Total_quantity_ordered: 
  SELECT Top 5
	dim_customers.customer_name,
    SUM(fact_order_lines.order_qty) AS Total_order_qty
 FROM dim_customers 
 INNER JOIN fact_order_lines ON dim_customers.customer_id = fact_order_lines.customer_id 
 GROUP BY dim_customers.customer_name
 ORDER BY Total_order_qty DESC;
```

### **Explanation:**

This query identifies the top 5 customers based on the total quantity of orders they have placed. It achieves this by connecting the `fact_order_lines` table with the `dim_customers` table using the `customer_id` field as a Primary key. The results are grouped by `customer_name`, and the `order_qty` values are summed to compute the overall ordered quantity for each customer. The outcomes are then sorted in descending order of `total_qty_ordered` and limited to the top 5 customers.

```sql
-- Top 5 Cusotmers by in_full_qty_ordered
 SELECT Top 5
	dim_customers.customer_name,
    SUM(fact_order_lines.delivery_qty) AS Full_qty_ordered
FROM dim_customers
INNER JOIN fact_order_lines ON dim_customers.customer_id = fact_order_lines.customer_id
GROUP BY dim_customers.customer_name
ORDER BY Full_qty_ordered DESC;   
```

### **Explanation:**

Similar to the query above, this one also seeks the top 5 customers, but this time by focusing on the in-full quantity of orders. It replaces the summation of `order_qty` with the summation of `delivery_qty` to calculate the in-full quantity ordered for each customer. The results are ordered by the `in_full_qty_ordered` field in descending order and limited to the top 5 customers.

```sql
-- Top 5 Customers by "OTIF" ordered Quantity.
WITH otif_ordered_qty AS (
	SELECT 
		fact_order_lines.customer_id,
        SUM(CASE WHEN fact_orders_aggregate.otif = 1 THEN fact_order_lines.delivery_qty ELSE 0 END) AS OTIF_Qty
    FROM fact_order_lines 
    INNER JOIN fact_orders_aggregate ON fact_order_lines.order_id = fact_orders_aggregate.order_id 
    GROUP BY fact_order_lines.customer_id
)
SELECT Top 5
	dim_customers.customer_name,
    otif_ordered_qty.OTIF_Qty
FROM otif_ordered_qty
INNER JOIN dim_customers ON otif_ordered_qty.customer_id = dim_customers.customer_id
ORDER BY OTIF_Qty DESC;
```

### **Explanation:**

<span style="color: var(--vscode-foreground);">This query identifies the top 5 customers based on the quantity of orders that were both on-time and in-full. It begins by creating a CTE named </span> `otif_ordered_qty`<span style="color: var(--vscode-foreground);">, which joins the </span> `fact_order_lines` <span style="color: var(--vscode-foreground);">table with the </span> `fact_orders_aggregate` <span style="color: var(--vscode-foreground);">table using the </span> `order_id` <span style="color: var(--vscode-foreground);">field. The results are grouped by </span> `customer_id`<span style="color: var(--vscode-foreground);">, and the </span> `delivery_qty` <span style="color: var(--vscode-foreground);">is summed only for records where the </span> `otif` <span style="color: var(--vscode-foreground);">field is equal to 1 (indicating on-time and in-full orders). This subquery is then joined with the </span> `dim_customers` <span style="color: var(--vscode-foreground);">table using the </span> `customer_id` <span style="color: var(--vscode-foreground);">field to obtain customer names. The outcomes are sorted in descending order of </span> `OTIF_Qty` <span style="color: var(--vscode-foreground);">and limited to the top 5 customers.</span>

### **Calculate the actual OT%, IF%, and OTIF% for each customer.**

```sql
WITH actual AS (
	SELECT 
		dim_customers.customer_name,
        SUM(CASE WHEN fact_orders_aggregate.on_time = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(DISTINCT fact_orders_aggregate.order_id) AS  actual_ot,
        SUM(CASE WHEN fact_orders_aggregate.in_full = 1 THEN 1 ELSE 0 END) * 100.0 / COUNT(DISTINCT fact_orders_aggregate.order_id) AS actual_if,
        SUM(CASE WHEN fact_orders_aggregate.otif = 1 THEN 1 ELSE 0 END) * 100.0/ COUNT(DISTINCT fact_orders_aggregate.order_id) AS actual_otif
	FROM fact_orders_aggregate
    JOIN dim_customers ON fact_orders_aggregate.customer_id = dim_customers.customer_id
    GROUP BY dim_customers.customer_name 
)
SELECT 
	actual.customer_name,
    ROUND(actual.actual_ot, 2) AS ot_pct,
    ROUND(actual.actual_if, 2) AS if_pct,
    ROUND(actual.actual_otif,2) AS otif_pct
FROM actual
ORDER BY actual.customer_name;
```

### **Explanation:**

This query calculates the actual percentages of on-time (OT%), in-full (IF%), and on-time and in-full (OTIF%) deliveries for each customer. It achieves this by retrieving data from the `fact_orders_aggregate` table and joining it with the `dim_customers` table based on the `customer_id` column.

Using a combination of SUM, COUNT, and CASE statements, the query determines the actual OT%, IF%, and OTIF% for each customer. The SUM function tallies the number of orders that meet the criteria for being on-time, in-full, or both (as indicated by specific columns in the `fact_orders_aggregate` table). Meanwhile, the COUNT function calculates the total number of orders per customer.

To calculate the actual OT%, IF%, and OTIF%, the query divides the sum of corresponding orders by the total number of orders for each customer. This value is then multiplied by 100 to express it as a percentage. The results are grouped by the `city` column in the `dim_customers` table, allowing for separate percentage calculations for each customer within each city.

In summary, the query generates the actual percentages for on-time, in-full, and on-time and in-full deliveries for each customer, revealing their respective performance metrics.

### **Identify the most and least ordered product by each customer.**

```sql
WITH customer_products AS (
    SELECT 
        dim_customers.customer_name,
        dim_products.product_name,
        COUNT(fact_order_lines.product_id) AS Product_count
    FROM fact_order_lines 
    JOIN dim_customers ON fact_order_lines.customer_id = dim_customers.customer_id
    JOIN dim_products ON fact_order_lines.product_id = dim_products.product_id
    GROUP BY dim_customers.customer_name, dim_products.product_name
),
customer_max_min_counts AS (
    SELECT
        cp.customer_name,
        MAX(cp.Product_count) AS max_product_count,
        MIN(cp.Product_count) AS min_product_count
    FROM customer_products cp
    GROUP BY cp.customer_name
)
SELECT
    cp.customer_name,
    MAX(CASE WHEN cp.Product_count = cmc.max_product_count THEN cp.product_name END) AS most_ordered_product,
    MAX(CASE WHEN cp.Product_count = cmc.min_product_count THEN cp.product_name END) AS least_ordered_product
FROM customer_products cp
JOIN customer_max_min_counts cmc ON cp.customer_name = cmc.customer_name
GROUP BY cp.customer_name, cmc.max_product_count, cmc.min_product_count
ORDER BY cp.customer_name;
```

### **Explanation:**

1\. The first CTE named `customer_products`, t<span style="color: var(--vscode-foreground);">his CTE retrieves information about each customer's interaction with different products. It calculates the count of products ordered by each customer, grouping the results by customer name and product name.</span>

2\. The second CTE named `customer_max_min_counts`<span style="color: var(--vscode-foreground);">,</span>  t<span style="color: var(--vscode-foreground);">his CTE calculates the maximum and minimum product counts for each customer, determined from the customer_products CTE. It aggregates the data from customer_products, computing the highest and lowest product counts for each customer.</span>

3\. Main SELECT statement: <span style="color: var(--vscode-foreground);">This final SELECT statement combines the data from both CTEs to find the most and least ordered products for each customer. It leverages the customer_products CTE and joins it with the customer_max_min_counts CTE, matching customers based on their names. The SELECT statement employs two MAX(CASE...) functions to identify the product names corresponding to the maximum and minimum product counts for each customer.</span>

The resulting output showcases the customer name, along with the most ordered product and the least ordered product for each customer, based on their respective counts. The final presentation is organized by customer name.

### **Distribute the total product orders by their categories and calculate the percentage share. Also, identify the top and worst selling products for each city.**

```sql
WITH city_categories AS 
(
    SELECT 
        dim_customers.city,
        dim_products.product_name,
        dim_products.category,
        COUNT(fact_order_lines.order_id) AS total_orders
    FROM fact_order_lines 
    JOIN dim_customers ON fact_order_lines.customer_id = dim_customers.customer_id
    JOIN dim_products ON fact_order_lines.product_id = dim_products.product_id
    GROUP BY dim_customers.city, dim_products.product_name, dim_products.category 
),
categories_totals AS 
(
    SELECT 
        city,
        SUM(CASE WHEN category = 'dairy' THEN total_orders ELSE 0 END) AS dairy_total,
        SUM(CASE WHEN category = 'food' THEN total_orders ELSE 0 END) AS food_total,
        SUM(CASE WHEN category = 'beverages' THEN total_orders ELSE 0 END) AS beverages_total,
        SUM(total_orders) AS total_orders 
    FROM city_categories
    GROUP BY city
)
SELECT 
    cc.city,
    cc.category,
    SUM(cc.total_orders) AS total_orders,
    CONCAT(ROUND((SUM(cc.total_orders) * 100.0 / ct.total_orders), 2), '%') AS percent_share,
    (SELECT TOP 1 cc2.product_name FROM city_categories cc2 WHERE cc2.city = cc.city ORDER BY cc2.total_orders DESC) AS top_selling_products,
    (SELECT TOP 1 cc2.product_name FROM city_categories cc2 WHERE cc2.city = cc.city ORDER BY cc2.total_orders ASC) AS least_selling_products
FROM city_categories cc
JOIN categories_totals ct ON cc.city = ct.city
GROUP BY cc.city, cc.category, ct.total_orders
ORDER BY cc.city, percent_share DESC;
```

### **Explanation:**

The first CTE named `city_categories`, in this CTE, data is extracted to compile information about total orders for specific product categories in each city. The query joins data from the `fact_order_lines`, `dim_customers`, and `dim_products` tables, grouping results by city, product name, and category. The SUM(CASE...) functions segment the total orders by category type. The final calculated `total_orders` field represents the grand total of orders for each city.

The final SELECT statement (the main statement) combines data from both CTEs to provide a comprehensive overview of product category distribution and sales performance in various cities. It calculates the total orders and the percentage share of each category's orders in relation to the overall orders for each city.

Additionally, the query employs subqueries to determine the top-selling and least-selling products for each city. These subqueries use the city\_categories CTE and ORDER BY clauses to find the product names with the highest and lowest total orders for each city.

The output of this query presents a breakdown of product category distribution for each city, showcasing the total orders, percentage share, top-selling products, and least-selling products. The results are organized by city and sorted based on the percentage share in descending order.
